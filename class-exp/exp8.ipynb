```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

class Doctor(Person):
    def __init__(self, doc_id, name, age, gender, specialty):
        super().__init__(name, age, gender)
        self.doc_id = doc_id
        self.specialty = specialty

    def display(self):
        print(f"ID: {self.doc_id} | Name: {self.name} | Specialty: {self.specialty}")

class Patient(Person):
    def __init__(self, pat_id, name, age, gender, ailment):
        super().__init__(name, age, gender)
        self.pat_id = pat_id
        self.ailment = ailment

    def display(self):
        print(f"ID: {self.pat_id} | Name: {self.name} | Ailment: {self.ailment}")

# Simple Management Controller
doctors = []
patients = []

def main_menu():
    while True:
        print("\n--- HOSPITAL MANAGEMENT SYSTEM ---")
        print("1. Add Doctor")
        print("2. Add Patient")
        print("3. View All Doctors")
        print("4. View All Patients")
        print("5. Exit")
        
        choice = input("Select an option: ")

        if choice == '1':
            d_id = input("Enter Doctor ID: ")
            name = input("Enter Name: ")
            age = input("Enter Age: ")
            gender = input("Enter Gender: ")
            spec = input("Enter Specialty: ")
            doctors.append(Doctor(d_id, name, age, gender, spec))
            print("Doctor added successfully!")

        elif choice == '2':
            p_id = input("Enter Patient ID: ")
            name = input("Enter Name: ")
            age = input("Enter Age: ")
            gender = input("Enter Gender: ")
            ailment = input("Enter Ailment: ")
            patients.append(Patient(p_id, name, age, gender, ailment))
            print("Patient added successfully!")

        elif choice == '3':
            print("\n--- List of Doctors ---")
            for d in doctors: d.display()

        elif choice == '4':
            print("\n--- List of Patients ---")
            for p in patients: p.display()

        elif choice == '5':
            print("Exiting System...")
            break
        else:
            print("Invalid Choice. Please try again.")

if __name__ == "__main__":
    main_menu()
```
