Import Libraries

```python
import pandas as pd
import numpy as np
```

Load the dataset

```python
df = pd.read_csv("student_data.csv") 
df.head(5)
```

1. Check for missing values

```python
print(df.isnull().sum())
```

2. Remove Rows with Missing Data ---------->

```python
# Drops any row that has at least one empty cell
df_clean = df.dropna()
df_clean.head(5)
```

3. Remove Duplicate Rows ---------->

```python
# Removes exact duplicate rows
df_clean = df_clean.drop_duplicates()
df.head(5)
```

4. Convert Binary Text to Numbers (0 and 1) ---------->

```python
# Change Yes to 1 and No to 0
df_clean['Internet_Access'] = df_clean['Internet_Access'].map({'Yes': 1, 'No': 0})
df_clean['Extracurricular_Activities'] = df_clean['Extracurricular_Activities'].map({'Yes': 1, 'No': 0})
df_clean.head(3)
```

5. Convert All Other Text to Numbers ---------->

```python
# This is the "magic" command for beginners to handle text data
df_final = pd.get_dummies(df_clean)
df_final.head(5)
```
